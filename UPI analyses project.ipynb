Weekend vs Weekday: Success Rate of UPI Transactions

```python
successful_transactions = df[df['transaction_status']=='SUCCESS'].groupby('is_weekend').size()
total_transactions = successful_transactions.sum()
successful_transactions_percentage = (successful_transactions/total_transactions)*100
successful_transactions_percentage
```

Transaction Success Rate by Sender State (%)

```python
ts_success = df[df['transaction_status']=='SUCCESS'].groupby('sender_state').size()
total_ts = df.groupby('sender_state').size()
ts_success_percentage = (ts_success/total_ts)*100
ts_success_percentage
```

Top Bank by Number of Transactions in Each Sender's State

```python
df.groupby('sender_state')['sender_bank'].value_counts().groupby(level=0).head(1).sort_values(ascending=False)
```

```python
df.groupby('sender_state')['receiver_bank'].value_counts().groupby(level=0).head(1).sort_values(ascending=False)
```
