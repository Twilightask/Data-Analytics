# Replacing blanks with 0 as they were casuing trouble in analysis

```python
df['TotalCharges'] = df['TotalCharges'].replace(' ','0')
```

# Changing the type of TotalCharges column from object to float

```python
df['TotalCharges'] = df['TotalCharges'].astype('float')
```

# Converting SeniorCitizen table values where 1 return Yes and where 0 return No

```python
def conv(value):
    if value == 1:
        return 'Yes'
    else:
        return 'No'
    
df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)    

```

```python
ax = sns.countplot(x = 'Churn', data = df)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
plt.figure(figsize=(4,4))
gb = df.groupby('Churn').agg({'Churn':'count'})
gb
plt.pie(gb['Churn'], startangle= 90, labels = gb.index, autopct= '%1.1f%%')
plt.title("Customer churn percentage")
plt.show()
```

# From the above pie chart we can see that 26.5% customer had churned out. Now  exploring the reason behind this

```python
ay = sns.countplot(x = 'gender', data = df, hue = 'Churn')
for container in ay.containers:
    ay.bar_label(container)
plt.title("Customer churn by Gender", fontsize = 10)
plt.show()
```

```python
az = sns.countplot(x = 'SeniorCitizen', data = df, hue = 'Churn')
for container in az.containers:
    az.bar_label(container)
```

```python
sns.histplot(x = 'tenure', data = df, hue = 'Churn', bins = 72)
plt.title("Customer churn data by tenure")
plt.show()
```

```python
# Create subplots (3 rows, 3 columns)
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 12))
fig.suptitle("Countplots for Different Services", fontsize=16)

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Creating countplots for each individual column
sns.countplot(x=df['PhoneService'], hue = df['Churn'], ax=axes[0])
axes[0].set_title("PhoneService")

sns.countplot(x=df['MultipleLines'],hue = df['Churn'], ax=axes[1])
axes[1].set_title("MultipleLines")

sns.countplot(x=df['InternetService'],hue = df['Churn'], ax=axes[2])
axes[2].set_title("InternetService")

sns.countplot(x=df['OnlineSecurity'],hue = df['Churn'], ax=axes[3])
axes[3].set_title("OnlineSecurity")

sns.countplot(x=df['OnlineBackup'],hue = df['Churn'], ax=axes[4])
axes[4].set_title("OnlineBackup")

sns.countplot(x=df['DeviceProtection'],hue = df['Churn'], ax=axes[5])
axes[5].set_title("DeviceProtection")

sns.countplot(x=df['TechSupport'],hue = df['Churn'], ax=axes[6])
axes[6].set_title("TechSupport")

sns.countplot(x=df['StreamingTV'],hue = df['Churn'], ax=axes[7])
axes[7].set_title("StreamingTV")

sns.countplot(x=df['StreamingMovies'],hue = df['Churn'], ax=axes[8])
axes[8].set_title("StreamingMovies")

# Adjust layout to prevent overlapping
plt.tight_layout(rect=[0, 0, 1, 0.96])  
plt.show()
```
